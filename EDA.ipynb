```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = pd.read_csv('Iris.csv')
data.head()
```

```python
data.info()
```

```python
sample = data.sample(n=5,replace=False,random_state=66)
sample
```

```python
# Scatter plot

# Simple create axes
ax = plt.axes()
# Label axes
ax.set(xlabel='LengthCm',ylabel='WidthCm',title='Length x Width')
sns.scatterplot(data=data,x='SepalLengthCm',y='SepalWidthCm',c = 'r')

sns.scatterplot(data=data,x='PetalLengthCm',y='PetalWidthCm',c = 'b')
```

```python
sns.histplot(data = data.PetalLengthCm,bins=25)
plt.title('Histplot of PetalLengthCm')
```

```python
fig, ax = plt.subplots()

ax.barh(np.arange(10), data.SepalWidthCm.iloc[:10])
ax.set_yticks(np.arange(np.arange(10)))
ax.set_yticklabels(np.arange(1,11))
```

```python
data.columns
```

```python
ax = data.iloc[:,1:].groupby(by = 'Species').mean()
ax.plot(color = ['r','b','g','y'],figsize=(6,4),fontsize=15)
```

```python
sns.set_context('talk')
sns.pairplot(data,hue='Species')
```

```python
sns.jointplot(x=data['SepalLengthCm'],y=data['SepalWidthCm'],kind='hex')
```

```python
# FaceGrid
plot = sns.FacetGrid(data,col='Species',margin_titles=True)
plot.map(plt.hist,'SepalLengthCm')
```
