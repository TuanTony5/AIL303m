```python
import pandas as pd
import numpy as np
import seaborn as sns
import math
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Telco_customer_churn.xlsx')
df.head()
```

```python
df.columns
```

```python
df['Payment Method'].value_counts()
```

```python
sns.barplot(y="Churn Value", x="Payment Method", data=df,errorbar=None)
plt.ylabel("Churn Rate")
plt.xlabel("Payment Type")
plt.title("Churn rate by payment type")
```

```python
sns.barplot(y="Churn Value", x=pd.cut(df['Tenure Months'],bins=5), data=df,errorbar=None)
plt.ylabel("Churn Rate")
plt.xlabel("Tenure range in Months")
plt.title("Churn rate by Tenure, Phone customer")
```
